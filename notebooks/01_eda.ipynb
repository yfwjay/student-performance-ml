```python
plt.style.use('ggplot')
student_df.hist(figsize =(16 , 10) , bins = 20 , edgecolor = 'black')
plt.suptitle('Feature Distribution' , fontsize = 22)
plt.tight_layout()
plt.show()
```

```python
# NUMERICAL VARIABLES

# studytime weekly - normal or evenly distributed study habit patterns
# absences - wide distribution with some higher values 
# age - evenly distributed between the ages of 15 and 18
# gpa - Approximately its normal as the gpa ranges around the mid levels

# CATEGORICAAL VARIABLES
# gender - even distribution
# ethinicity - One ethnic group dominates while the remaining ethnic groups they are evenly distributed
# parentaleducation - Most parents have mid level education
# tutoring - uneven distribution. More 0 than ones


# ACTIVITY VARIABLES

# extracurricular - people who dont participate are more than those who participate
# music - most people dont participate as there are more 0 than ones(strong imbalance)
# sports - most people dont participate as there are more 0 than ones(strong imbalance)
# volunteering - uneven distribution as one categorey dominates (0)

# THE TARGETS(grade class)

# gradeclass - There is imbalance as one category dominates mostly(4) . This would afffect our models so it would require as to do so some refactoring before feeding it into our model
```

```python
# We can draw a countplot for one to see that the grade class we have an imbalance we see the count of each of the classes

fig , axes = plt.subplots(figsize = (12 , 6))

sns.countplot(ax = axes , x = 'GradeClass' , data = student_df)

axes.set_title('A Count Plot showing the Distribution of the classes' , fontsize = 22)
```

```python
# Heatmap
corr = student_df.select_dtypes(include = 'number').corr()

fig , axes = plt.subplots(figsize = (12 , 6))

sns.heatmap(corr, ax = axes ,  annot = True , fmt = '.2f' , cmap = 'coolwarm' , linewidths = 0.5 , linecolor = 'white' , cbar = True)

axes.set_title("A Heatmap showing the correlation between the columns" , fontsize = 22)
plt.tight_layout()
```

```python
# Explanation of the values of the heatmap

# Absences and GPA - Strong negative correlation. High absences results to a lower GPA

# Absences and Gradeclass - Strong postive correlation. Increase in the number of absences high gradeclass(poorer grade)

# Tutoring and GPA - Weak postive correlation. Tutoring doesn't contribute alot to GPA

# The same applies to (ParentalSupport ,Extracurricular , Sports , Music , Volunteering)

# GPA and Grade Class . Strong negative correlation . Higher gpa , results to a lower grade class(better performance)
```

```python
plt.figure(figsize = (12 , 8))

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm")
plt.tight_layout()
plt.title("A Heat map where we have masked the top righ side to reduce cognitive load" , fontsize = 22)
plt.show()
```

```python
# We need to map the gender column so that the numeric values in the gender column represent male and female

mapping = {0 : 'female' , 1 : 'Male'}  # create a dictionary with the values we want to change

# create a new column gender_label

student_df['gender_label'] = student_df['Gender'].map(mapping)
student_df['gender_label']
```

```python
bar_df = student_df.loc[: , ['GPA' , 'gender_label']]

bar_df.groupby('gender_label').mean()
```

```python
# SIMPLE VISUALISATIONS


# BARCHART - Gender and GPA

fig , axes = plt.subplots(figsize = (12 , 5))

bar_chart = axes.bar(student_df['gender_label'] , student_df['GPA'] , color = 'skyblue' , edgecolor = 'red' , label = 'Gender')

axes.set_title("Bar Chart showing Gender Against GPA")
axes.legend()
axes.set_xlabel("Gender")
axes.set_ylabel('GPA')
```

```python
# ExtraCurricular Activities and GPA


mapping = {0 : 'Yes' , 1 : 'No'}

student_df['ExtraCurricularLabel'] = student_df['Extracurricular'].map(mapping)


fig , axes = plt.subplots(figsize = (12 , 5))

bar_chart = axes.bar(student_df['ExtraCurricularLabel'] , student_df['GPA'] , color = 'skyblue' , edgecolor = 'red' , label = 'Extracurricular')

axes.set_title("Bar Chart showing Extracurricular Against GPA")
axes.legend()
axes.set_xlabel("Extracurricular")
axes.set_ylabel('GPA')
```
